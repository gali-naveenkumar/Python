```python
# importing Requried Lirbraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# importing data using seaborn library
# preprocessing data using pandas library
# visualising data using matplotlib.pyplot library
```

```python
# importing tips data from seaborn library
df = sns.load_dataset("tips")
df.head()
```

```python
type(df)
```

```python
print("Sum of Total Bill is:",df['total_bill'].sum())
```

```python
# Sum of total Bill by Each Gender
df.groupby("sex")['total_bill'].sum()
```

```python
df.groupby("sex")['total_bill'].sum().plot(kind='bar')
```

```python
plt.bar(df['sex'],df['total_bill'])
plt.show()
```

```python
# summrize the Gender wise sum of total bill using pandas
gws = pd.pivot_table(data=df,values='total_bill',columns='sex',aggfunc='sum')
gws
```

```python
# converting gws output from horizontal to vertical
gws = gws.T
gws
```

```python
gws.columns
```

```python
# getting the index values in a column
gws.reset_index(inplace=True)
gws
```

```python
gws.columns
```

```python
# to drop a column
df.drop(columns='column_name',inplace=True)
```

```python
# creating visual using matplotlib.pyplot library
plt.bar(gws['sex'],gws['total_bill'])
plt.show()
```

```python
# Histogram using pandas
df['total_bill'].plot(kind='hist')
```

```python
# Creating histogram using matplotlib library
plt.hist(df['total_bill'],color='black')
plt.title("Distribution of Bill!")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot is used to show the relation ship between two columns
plt.scatter(df['tip'],df['total_bill'],color='black')
plt.title("Tip vs TotalBill!")
plt.xlabel("Tip")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Scatter plot is used to show the relation ship between two columns
plt.scatter(df['size'],df['total_bill'],color='black')
plt.title("Tip vs TotalBill!")
plt.xlabel("Tip")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Box plot
plt.boxplot(df['total_bill'])
plt.show()
```
