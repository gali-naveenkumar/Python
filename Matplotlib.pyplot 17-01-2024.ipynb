# Matplotlib Library

```python
# Matplotlib is data Visualisation Library of Python
# It is used to create Low Level Graphs
# Bar, Column, Clusterd Column, Clusterd Bar, Pie, Line, BoxPlot, Histogram
# Scatter etc.
```

```python
# !pip install matplotlib
import matplotlib
```

```python
# to check the version
print(matplotlib.__version__)
```

```python
# importing pyplot module
import matplotlib.pyplot as plt
```

```python
# creating List to store Monthly sales
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
import random
sales = []
for i in range(1,13):
    r = random.randint(5000,45000)
    sales.append(r)
print(months)
print(sales)
```

```python
# Converting months into x and sales into y
x = months
y = sales
```

```python
# Line Chart is used to show the trend of the data
# line chart is used on continous data
# to create line chart use plot method of pyplot
# the plot function is used to draw point in a line chart
plt.plot(x,y)
plt.show()
```

```python
# adding title to the chart
plt.plot(x,y)
# chart title
plt.title("Montly Sales Trend!")
# x axis title
plt.xlabel("Months")
# y axis title
plt.ylabel("Sales")
plt.show()
```

```python
# Customizing the chart
plt.plot(x,y,marker='X')
plt.title("Montly Sales Trend!")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# Customizing the chart
plt.plot(x,y,marker='X',ms=15)
plt.title("Montly Sales Trend!")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# Customizing the chart
plt.plot(x,y,marker='X',ms=15,mec='black')
plt.title("Montly Sales Trend!")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# Customizing the chart
plt.plot(x,y,marker='X',ms=10,mec='black',mfc='white')
plt.title("Montly Sales Trend!")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# Customizing the chart
plt.plot(x,y,marker='X',ms=15,mec='black',mfc='white',color='green')
plt.title("Montly Sales Trend!")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# Customizing the chart
plt.plot(x,y,marker='X',ms=15,mec='black',mfc='white',color='green',linestyle='--')
plt.title("Montly Sales Trend!")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# Bar chart is used to show the distribution of the categorical data
plt.barh(x,y,color='orange')
plt.title("Comparision Sales By Month!")
plt.xlabel("Sales")
plt.ylabel("Months")
plt.show()
```

```python
# Column chart is used to show the distribution of the categorical data
plt.bar(x,y,color='orange')
plt.title("Comparision Sales By Month!")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
years = [2020,2021,2022]
y_sales = [150000,75000,210000]
```

```python
# Pie chart is used to show the distribution of categorical data
# Frequency and Relative Frequency
plt.pie(y_sales,labels=y_sales,explode=[0,0.2,0],autopct="%1.1f%%",shadow=True)
plt.legend(years)
plt.show()
```

```python
plt.pie(y,labels=y)
plt.legend(x)
plt.show()
```
