```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
# gender wise total bill
gws = pd.pivot_table(data=df,values='total_bill',
                    columns='sex',aggfunc='sum').T
gws.reset_index(inplace=True)
gws
```

```python
# day wise total bill
dws = pd.pivot_table(data=df,values='total_bill',
                     columns='day',aggfunc='sum').T
dws.reset_index(inplace=True)
dws
```

```python
plt.bar(gws['sex'],gws['total_bill'],color='green')
plt.title("Gender Wise Total Bill!")
plt.xlabel("Gender")
plt.ylabel("Total Bill")
plt.show()
```

```python
plt.pie(dws['total_bill'],labels=dws['total_bill'],autopct="%1.1f%%")
plt.title("Day Wise Total Bill")
plt.legend(dws['day'])
plt.show()
```

```python
# Subplot method of matplotlib.pyplot is used to draw multiple charts
# in one frame
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.bar(gws['sex'],gws['total_bill'],color='green')
plt.title("Gender Wise Total Bill!")
plt.xlabel("Gender")
plt.ylabel("Total Bill")

plt.subplot(1,2,2)
plt.pie(dws['total_bill'],labels=dws['total_bill'],autopct="%1.1f%%")
plt.title("Day Wise Total Bill")
plt.legend(dws['day'])
plt.show()
```

# Error handling in python

```python
# Error handling in python is used manipulate errors
# example
print(name)
```

```python
# four types of method in python to handle errors
# try method in python is used to test the code
# except method in python is used to handle the error
```

```python
try:
    print(name)
except:
    print("Hello")
```

```python
print("d"/5)
```

```python
try:
    print("d"/5)
except:
    print("Check the Equation!")
```

```python
try:
    num1 = int(input("Enter Numerator: "))
    num2 = int(input("Enter Dimnominator: "))
    print(num1/num2)
except:
    print("Plz Enter the Positive Number Only")
```

```python
# else method in error handling
# is used to execute code when there is no error 
try:
    num1 = int(input("Enter Numerator: "))
    num2 = int(input("Enter Dimnominator: "))
    print(num1/num2)
except:
    print("Plz Enter the Positive Number Only")
else:
    print("Code sucessfully Excecuted!")
```

```python
# Finally method in Error handling
try:
    num1 = int(input("Enter Numerator: "))
    num2 = int(input("Enter Dimnominator: "))
    print(num1/num2)
except:
    print("Plz Enter the Positive Number Only")
else:
    print("Code sucessfully Excecuted!")
finally:
    print("Done!")
```
